### Step 1: Data Scraping and Preparation


```python
import requests
from datetime import datetime
import pandas as pd
from bs4 import BeautifulSoup
html = requests.get('https://cmsc320.github.io/files/top-50-solar-flares.html').content
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
soup.prettify()
```

```python
result = soup.find(class_="table table-striped table-responsive-md")
table = pd.concat(pd.read_html(str(result)))
table.columns = ['rank', 'x_classification', 'date', 'region', 'start_time', 'maximum_time', 'end_time', 'movie']
table
```

```python
table.columns = ['rank', 'x_classification', 'date', 'region', 'start_time', 'maximum_time', 'end_time', 'movie']
table
```

#### I extracted the page using request on the url and then using Beautiful soup to parse the data starting from where the table is on the page. After getting all the information into a dataframe, I appropriately named each column.

### STEP 2: Tidying the Top 50 Solar Flare Data
#### Here I dropped the last column of the table, combined the date and each of the three time columns into three datetime columns and updated the values in the dataframe along the way. I set regions coded as "-" as missing, replacing them with 0.



```python
table = table.drop('movie', axis = 1)
table
```

```python
for index, row in table.head(n=50).iterrows():

    table.at[index,'start_time'] = pd.to_datetime(row['date'] + " " + row['start_time'])
    table.at[index,'maximum_time'] = pd.to_datetime(row['date'] + " " + row['maximum_time'])
    table.at[index,'end_time'] = pd.to_datetime(row['date'] + " " + row['end_time'])

#dropping the date column
table = table.drop('date', axis = 1) 

#renaming the start, maximum, and end columns
table.rename(columns = {'start_time':'start_datetime', 'maximum_time':'maximum_datetime', 'end_time':'end_datetime'}, inplace = True)
    
```

```python
#setting regions coded as "-" as a missing data value
table['region'] = table['region'].replace('-','0')


#changing position of region column
new_cols = ['rank', 'x_classification', 'start_datetime', 'maximum_datetime', 'end_datetime', 'region']
table = table[new_cols]
#cleaning up classification values
table = table.replace({'\+': '.0'}, regex=True)
table = table.replace({'X17.2.0': 'X17.2'}, regex = True) 
table
```

#### I tidied the data by combining the time columns with the date and then dropping the extra columns and also moved the region column. I then cleaned up the x classification values making sure that any value that was written like this "39+" was converted to a well defined float.  

## STEP 3: Scraping the NASA data
### Scraping the data from a mirror of the NASA website to get additional data about the solar flares.

```python
nasa_html = requests.get('https://cmsc320.github.io/files/waves_type2.html')
nasa_soup = BeautifulSoup(nasa_html.text, "lxml")
nasa_soup.prettify()
info = nasa_soup.find('pre').get_text()
sep = info.split('\n')
nasa_table = pd.DataFrame(sep)
for i in range(12):
    nasa_table = nasa_table.drop(i, axis = 0)

nasa_table.drop([530,531], axis = 0, inplace = True)
nasa_table = nasa_table.reset_index(drop=True)
nasa_table
```

```python
nasa_table = nasa_table[0].str.split(expand=True)
nasa_table.drop(nasa_table.loc[:, 15:23].columns, axis=1, inplace = True)
nasa_table.columns = ['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region', 'flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed', 'plot']
nasa_table
```

```python
nasa_table
```

## STEP 4
### Tidying the NASA table by coding missing observations properly, recoding columns that correspond to more than one piece of information, and treating dates and times appropriately.



```python
nasa_table = nasa_table.replace({'-----': 'NaN'}, regex=True)
nasa_table = nasa_table.replace({'----': 'NaN'}, regex=True)
nasa_table = nasa_table.replace({'---': 'NaN'}, regex=True)
nasa_table = nasa_table.replace({'--/--': 'NaN'}, regex=True)
nasa_table = nasa_table.replace({'--:--': 'NaN'}, regex=True)
```

```python
is_halo = []
for i in nasa_table['cme_angle']:
    if i == 'Halo':
        is_halo.append(True)
    else:
        is_halo.append(False)
        
nasa_table['is_halo'] = is_halo
nasa_table = nasa_table.replace({'Halo': 'NaN'}, regex=True)
```

```python
width_lower_bound = []
for i in nasa_table['cme_width']:
    if ">" in i:
        width_lower_bound.append(True)
    else:
        width_lower_bound.append(False)
nasa_table['width_lower_bound'] = width_lower_bound
 

nasa_table = nasa_table.replace({'>': ''}, regex=True)
```

```python
for index, row in nasa_table.iterrows():
    year = row['start_date'][0:4] + "/"
    nasa_table.at[index, 'end_date'] = year + row['end_date']
    nasa_table.at[index, 'cme_date'] = year + row['cme_date']
    nasa_table.at[index,'start_time'] = pd.to_datetime(row['start_date'] + " " + row['start_time'])
```

```python
nasa_table['end_time'] = nasa_table['end_time'].str.replace('24:00','23:59')
nasa_table['cme_time'] = nasa_table['cme_time'].str.replace('24:00','23:59')
for index, row in nasa_table.iterrows():
    nasa_table.at[index,'end_time'] = pd.to_datetime(row['end_date'] + " " + row['end_time'],)
    nasa_table.at[index,'cme_time'] = pd.to_datetime(row['cme_date'] + " " + row['cme_time'], errors = 'coerce')
```

```python
nasa_table = nasa_table.drop('start_date', axis = 1) 
nasa_table = nasa_table.drop('cme_date', axis = 1) 
nasa_table = nasa_table.drop('end_date', axis = 1) 
nasa_table.rename(columns = {'start_time':'start_datetime', 'end_time':'end_datetime', 'cme_time':'cme_datetime'}, inplace = True)
#cleaning up flare classsification values
for index, row in nasa_table.iterrows():
    if row['flare_classification'][-1] == ".":
        nasa_table.at[index, 'flare_classification'] = row['flare_classification'] + "0"
        
nasa_table
```

## PART 2: Analysis

#### Here I replicate the top 50 solar flare table in SpaceWeatherLive.com exactly using the data obtained from NASA 

```python
#putting the table in order
top50 = nasa_table.sort_values('flare_classification', ascending = False)
top50 = top50.reset_index(drop=True) 
#only viewing the top 50 flares
top50 = top50[:50]
#Taking away 'X' and converting the flare_classification column to a float to acccurately order the dataframe
top50['flare_classification'] = top50['flare_classification'].str.replace('X','').astype(float)
top50 = top50.sort_values('flare_classification', ascending = False)
top50 = top50.reset_index(drop=True) 
#Reinserting 'X' after sorting is complete
top50['flare_classification'] = 'X' + top50['flare_classification'].astype(str)
#Dropping unnecessary columns 
top50 = top50.drop(['end_frequency', 'flare_location', 'cme_angle', 'cme_width', 'cme_speed', 'plot', 'is_halo', 'width_lower_bound'], axis = 1) 

top50['start_datetime'] = pd.to_datetime(top50['start_datetime'])

top50
```

### REPLICATION
#### After ordering all solar flares from NASA and only looking at the top 50, some of the information doesnt match up. For example, the flare reported by NASA which has X9.4 classification started on 1997-04-01 14:00:00 while the Spaceweather live flare that matched up to this one(region 8100) had 1997-11-06 11:49:00. So the data is not exactly the same for the dates and times. 


### INTEGRATION
#### I decided to first match by xclassification because only 8 flares were left out compared to the 14 flares that were left out when I tried to match by the dates first, and over 40 flares were left unmatched when trying to match by region first. Doing it this way, I was able to match almost every single flare and the only ones left were flares that didn't match in any way between the dataframes. When multiple flares had the same classification, I compared the flares by their dates. By doing this, I was able to match each flare without having multiple best matches.

### ANALYSIS
#### Here I create a plot that shows the top 50 solar flares in context with all data available in the NASA dataset.

```python
swl_table = table
def match(row):
    #matching flare_classification first
    matched_flares = top50.loc[top50['flare_classification'] == row['x_classification']]
    if len(matched_flares) == 1:
        top50.loc[matched_flares.index, 'rank'] = row['rank']
    elif len(matched_flares) > 1:
        #looking for a single date match when multiple flares have the same classification 
        #all flares in top50 have different dates so matched_date will never be greater than 1
        matched_date = matched_flares.loc[matched_flares['start_datetime'].dt.date == row['start_datetime'].date()]
        #flares have been narrowed down to the same classification and date
        top50.loc[matched_date.index, 'rank'] = row['rank']         
            
    else: 
        #matching date first
        matched_date = top50.loc[top50['start_datetime'].dt.date == row['start_datetime'].date()]
        if len(matched_date) == 1: 
            top50.loc[matched_date.index, 'rank'] = row['rank']
        else: 
            matched_region = top50.loc[top50['flare_region'] == row['region']]
            if len(matched_region) == 1:
                print("There was an exact match for a region after no match for flare or date")
                top50.loc[matched_region.index, 'rank'] = row['rank']
            
        
for index, row in swl_table.iterrows():
    match(row)
                             
top50                            
    
```

```python
import matplotlib.pyplot as plt
#converting the type to int to make sure start_frequency shows up in the correct order on the y axis
nasa_table = nasa_table.replace({'????': '0000'})
nasa_table['start_frequency'] = nasa_table['start_frequency'].astype(int)
top50['start_frequency'] = top50['start_frequency'].astype(int)
                                                                
plt.rcParams["figure.figsize"] = (8,5.5)

plt.scatter(nasa_table['start_datetime'],nasa_table['start_frequency'], c = 'green')
plt.scatter(top50['start_datetime'],top50['start_frequency'], c = 'red')

plt.xlabel("Year of Flare")
plt.ylabel("Start Frequency")
plt.title("Start Frequency of Solar Flares From 1996 to 2020")
```
